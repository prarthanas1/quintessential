```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dd = pd.read_excel(r"C:\Users\Prarthana Saikia\Desktop\Praxis\Machine Learning\Store data.xlsx")
dd.head()
```

```python
dd['cnt']=1
groups=dd.groupby(['CASH_MEMO','PRODUCT'])['cnt'].sum()
```

```python
groups_unstack=dd.groupby(['CASH_MEMO','PRODUCT'])['cnt'].sum().unstack().fillna(0)

def recode(x):
    if x<=0:
        return 0
    if x>0:
        return 1
    
basket_bin = groups_unstack.applymap(recode)
basket_rules = apriori(basket_bin,min_support=0.01, use_colnames=True)
basket_rules
```

```python
association_rules(basket_rules, metric='lift', min_threshold=2).sort_values(by='lift', ascending=False)
```
