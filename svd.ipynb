```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=plt.imread("C:/Users/DELL/Desktop/download.jpg")
plt.imshow(img)
img.shape
```

```python
img_new=img[:,:,0]
plt.imshow(img_new)
img_new.shape
```

```python
from scipy.linalg import svd
U, s, V=svd(img_new)
```

```python
import numpy as np
img_new_1=np.dot(np.dot(U[:,:10],np.diag(s[:10])),V[:10,:])
plt.imshow(img_new_1)
```

```python
img_new_2=np.dot(np.dot(U[:,:40],np.diag(s[:40])),V[:40,:])
plt.imshow(img_new_2)
```
